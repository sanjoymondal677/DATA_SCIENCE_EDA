```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/customer-shopping-trends-dataset/shopping_trends_updated.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
import missingno as misno
# misno.bar(data)
misno.matrix(data, figsize=(12, 6), sparkline=False, labels=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Gender',y='Purchase Amount (USD)',data=data,color='#E1AA74')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Purchase Amount (USD)',y='Age',hue='Gender',data=data)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.catplot(x='Gender',y='Age', kind='bar', data=data, palette="Set2", height=4, aspect=1.5, legend=False)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Item Purchased', y='Purchase Amount (USD)', data=data,palette="Set2")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(18,7))
sns.barplot(x='Location',y='Purchase Amount (USD)',data=data,hue='Gender',)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Payment Method',data=data,palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Subscription Status',data=data,hue='Gender',palette='mako')
plt.show()
```

```python
sns.countplot(x='Frequency of Purchases',data=data,hue='Gender')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='Item Purchased',data=data,hue='Gender',palette='YlOrBr',saturation=.75, width=.8)
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Frequency of Purchases', y='Purchase Amount (USD)', data=data, marker='o', color='b', label='Sales Trend')
plt.show()
```
