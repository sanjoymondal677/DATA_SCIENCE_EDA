```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/kaggle/input/top1000youtubers/youtubers_df.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.sort_values('Rank')
```

<h1>
Analyis only top 10 rank youtubs channel ðŸ˜„
    </h1>

```python
data = data.loc[data['Rank']<=10]
```

```python
plt.figure(figsize=(14,5))
ax = sns.barplot(x='Username',y='Suscribers',data=data,color='#B0D9B1')
for labels in ax.containers:
    ax.bar_label(labels)
plt.xticks(rotation = 30)
```

```python
plt.figure(figsize=(14,7))
sns.countplot(x='Country',data=data,color='#186F65')
plt.show()
```

```python
plt.figure(figsize=(14,5))
ax = sns.barplot(x='Username',y='Visits',data=data,color='#E5CFF7')
for labels in ax.containers:
    ax.bar_label(labels)
plt.xticks(rotation = 30)
```

```python
plt.figure(figsize=(14,5))
ax = sns.barplot(x='Username',y='Likes',data=data,color='#3085C3')
for labels in ax.containers:
    ax.bar_label(labels)
plt.xticks(rotation = 30)
```

```python
plt.figure(figsize=(14,5))
ax = sns.barplot(x='Username',y='Comments',data=data,palette = 'mako')
for labels in ax.containers:
    ax.bar_label(labels)
plt.xticks(rotation = 30)
```

```python
drop_null = data.dropna(subset=['Categories'])
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from wordcloud import ImageColorGenerator
text = " ".join(i for i in drop_null.Categories)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="#FFCF9D").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
