```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/amazon-books/Amazon Books Data.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
import missingno as misno
misno.bar(data)
plt.show()
```

```python
null_columns = data.columns[data.isnull().all()]
```

```python
data = data.drop(columns=null_columns)
```

```python
import missingno as misno
misno.bar(data)
plt.show()
```

```python
data = data.dropna()
```

```python
import missingno as misno
misno.bar(data)
plt.show()
```

```python
# data['rating'] = data['rating'].replace('$','').strip()astype(float)

# data['rating'] = data['rating'].replace('$', '')
```

```python
# Clean and convert the 'rating' column to float
data['rating'] = data['rating'].str.replace('[^\d.]', '', regex=True).astype(float)
```

```python
data.info()
```

```python
sorted_data = data.sort_values(by='rating',ascending=False)
top_10_getting_rating_author_name = sorted_data.head(10)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='author',y='rating',data=top_10_getting_rating_author_name,palette='Blues')
for label in ax.containers:
    ax.bar_label(label)
plt.xticks(rotation = 34)
plt.title("TOP 10 AUTHORS NAME WITH HIGHEST RATING")
plt.xlabel("AUTHORS_NAME",fontsize=20)
plt.ylabel("RATINGS",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='author',y='rating',data=top_10_getting_rating_author_name,color='#D988B9',hue='review_count')
for label in ax.containers:
    ax.bar_label(label)
plt.xticks(rotation = 34)
plt.title("TOP 10 AUTHORS NAME WITH HIGHEST RATING")
plt.xlabel("AUTHORS_NAME",fontsize=20)
plt.ylabel("RATINGS",fontsize=20)
plt.show()
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from wordcloud import ImageColorGenerator
text = " ".join(i for i in data.description)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
data['price'] = data['price'].str.replace('[^\d.]', '', regex=True).astype(float)
```

```python
data['price'].max()
```

```python
sorted_data_for_price = data.sort_values(by='price',ascending=False)
top_10_highest_price = sorted_data_for_price.head(10)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='title',y='price',data=top_10_highest_price,palette='dark:#5A9_r')
for label in ax.containers:
    ax.bar_label(label)
plt.xticks(rotation = 34)
plt.title("TOP 10 Books Price")
plt.xlabel("Title",fontsize=20)
plt.ylabel("Price",fontsize=20)
plt.show()
```
