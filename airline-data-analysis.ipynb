```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/kaggle/input/airline-dataset/Airline Dataset Updated - v2.csv")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Full_Name'] = data['First Name'] +" "+data['Last Name']
```

```python
ax=sns.countplot(data=data,x='Gender')
for labels in ax.containers:
    ax.bar_label(labels)
plt.show()
```

```python
age_data = data.loc[data['Age']>=60]
plt.figure(figsize=(12,6))
ax = sns.countplot(data=age_data,x='Age',hue='Gender')
plt.show()
```
