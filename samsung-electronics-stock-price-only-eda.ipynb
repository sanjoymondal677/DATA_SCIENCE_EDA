```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../input/samsung-electronics-stock-historical-price/005930.KS.csv")
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.dtypes
```

```python
data.head()
```

```python
import plotly.express as px
fig = px.line(x=data["Date"], y=data["Open"], title='Open value in years')
fig.show()
```

```python
import plotly.express as px
fig = px.line(x=data["Date"], y=data["Close"], title='Close values over the years')
fig.show()
```

```python
import plotly.express as px
fig = px.line(x=data["Date"], y=data["Low"], title='Low values over the years')
fig.show()
```

```python
import plotly.express as px
fig = px.line(x=data["Date"], y=data["High"], title='High values over the years')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime


fig = go.Figure(data=[go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'])])

fig.show()
```

```python
plt.figure(figsize=(22,8))
sns.lineplot(x='Date',y='Low',color="red",label="Low",data=data)
sns.lineplot(x="Date",y="High",color='green',label="High",data=data)
plt.xlabel("Date",fontweight='bold',size=24)
plt.ylabel("Values",fontweight='bold',size=24)
plt.show()
```

```python
plt.figure(figsize=(20,9))
sns.barplot(x = data['Date'].dt.month,y = 'Volume',data=data,palette="viridis")
plt.show()
```
