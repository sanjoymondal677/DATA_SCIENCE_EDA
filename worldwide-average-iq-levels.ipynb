```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)
```

```python
data = pd.read_csv("/kaggle/input/worldwide-average-iq-levels/IQ_level.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

axs[0, 0].bar(data['country'], data['IQ'], color='skyblue')
axs[0, 0].set_title('Country and IQ')
axs[0, 0].set_xticklabels(data['country'], rotation=90)

axs[0, 1].bar(data['country'], data['education_expenditure'], color='orange')
axs[0, 1].set_title('Country and education_expenditure')
axs[0, 1].set_xticklabels(data['country'], rotation=90)

axs[1, 0].bar(data['country'], data['avg_income'], color='green')
axs[1, 0].set_title('Country and avg_income')
axs[1, 0].set_xticklabels(data['country'], rotation=90)

axs[1, 1].bar(data['country'], data['avg_temp'], color='red')
axs[1, 1].set_title('Country and avg_temp')
axs[1, 1].set_xticklabels(data['country'], rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.scatter(data['IQ'], data['education_expenditure'], color='blue', label='IQ and education_expenditure')
plt.scatter(data['IQ'], data['avg_income'], color='green', label='IQ withavg_income')
plt.scatter(data['IQ'], data['avg_temp'], color='red', label='IQ and avg_temp')

plt.xlabel('IQ')
plt.ylabel('education_expenditure and avg_income and avg_temp')
plt.title('Scatter Plot with Three Datasets')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='rank',y='country',data=data)
plt.show()
```
