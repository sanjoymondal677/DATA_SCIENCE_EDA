<h4> Categorical Data:</h4>
<h4>Bar Chart:</h4>
<p>Description: Used to display the counts or frequencies of categories.<br>
Visualization: Categories on the x-axis and their corresponding frequencies on the y-axis.</p>


```python
import matplotlib.pyplot as plt

# Sample data for the bar graph
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 50, 30, 45]

# Create a bar graph
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.bar(categories, values, color='skyblue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph Example')

# Display the bar graph
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Automatically adjust subplot parameters for better layout
plt.show()
```

<h2>Pie Chart:</h2>
<p>Description: Displays the proportion of each category out of a whole.<br>
Visualization: Circular chart divided into slices, each representing a category's proportion.</p>

```python
# Sample data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]  # Values should add up to 100 for a valid pie chart
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode the 1st slice (Category A)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.tight_layout()
plt.show()
```

<h4>Stacked Bar Chart:</h4>

<p>Description: Shows the composition of each category by stacking bars for different sub-categories.<br>
Visualization: Categories on the x-axis, and bars stacked for different sub-categories, displaying their compositions.</p>

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values_series1 = [20, 35, 30, 25]  # Data series 1
values_series2 = [10, 25, 20, 30]  # Data series 2

# Create a stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values_series1, label='Series 1', color='skyblue')
plt.bar(categories, values_series2, label='Series 2', color='orange', bottom=values_series1)  # Stacking on top of series 1

# Set labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')

# Show legend
plt.legend()

# Display the stacked bar chart
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

<h4>Stacked Area Chart:</h4>

<p>Description: Similar to stacked bar charts but uses areas to represent categories.<br>
Visualization: Time (or any other continuous dimension) on the x-axis, and areas stacked on top of each other, representing different categories.</p>

```python

# Sample data for the stacked area chart
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values_series1 = [20, 35, 30, 25]  # Data series 1
values_series2 = [10, 25, 20, 30]  # Data series 2
values_series3 = [15, 20, 25, 22]  # Data series 3

# Create a stacked area chart
plt.figure(figsize=(8, 6))
plt.stackplot(categories, values_series1, values_series2, values_series3, labels=['Series 1', 'Series 2', 'Series 3'], alpha=0.7)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Area Chart Example')

# Show legend
plt.legend()

# Display the stacked area chart
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

<h4>Donut Chart:</h4>

<p>Description: Similar to a pie chart but with a hole in the center.<br>
Visualization: Circular chart with a hole in the center, displaying proportions like a pie chart.</p>

```python
# Sample data for the donut chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]  # Values should add up to 100 for a valid pie chart
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode the 1st slice (Category A)

# Create a donut chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, wedgeprops={'edgecolor': 'gray'})
plt.gca().add_artist(plt.Circle((0,0),0.6,fc='white'))  # Add a circle in the center to create a donut shape

# Display the donut chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.tight_layout()
plt.show()
```
