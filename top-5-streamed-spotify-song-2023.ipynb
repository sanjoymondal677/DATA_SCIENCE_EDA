```python
#import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the dataset
data = pd.read_csv("/kaggle/input/top-spotify-songs-2023/spotify-2023.csv",encoding='latin-1')
```

```python
#print the data set upto 6 rows
data.head(6)
```

```python
#representation of missing columns as graphical
import missingno as mssno
mssno.bar(data)
plt.show()
```

<h4>
Here we can see two columns we have missing values, according to importness we can drop or fillnaüòÑ.</h4>

```python
#convert streams column object ot float or int type
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')
```

```python
data.dtypes
```

```python
data['streams'].value_counts
```

```python
data['track_name'].value_counts
```

<h3>
    Top 5 streams, track with their artist name. üòç 
    </h3>

```python
sorted_df = data.sort_values(by='streams',ascending=False)
top_five_rows = sorted_df.head(5)
plt.figure(figsize=(10,8))
ax = sns.barplot(x='track_name',y='streams',data=top_five_rows,color='#A2C579',hue='artist(s)_name')
for label in ax.containers:
    ax.bar_label(label)
plt.xticks(rotation=32)
plt.show()
```
